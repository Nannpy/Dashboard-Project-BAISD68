```python
import pandas as pd

df = pd.read_csv('hydroponics_data.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index('timestamp').sort_index()
```

```python
numeric_cols = ['pH', 'TDS', 'water_level', 'DHT_temp', 'DHT_humidity', 'water_temp']
categorical_cols = ['pH_reducer', 'add_water', 'nutrients_adder', 'humidifier', 'ex_fan']
```

```python
df_numeric = df[numeric_cols].resample('5min').mean().round(3)
df_cat = df[categorical_cols].resample('5min').last()
```

```python
df_resampled = pd.concat([df_numeric, df_cat], axis=1).reset_index()
df_resampled
```

```python
df_resampled.isna().sum()
```

```python
df_resampled = df_resampled.dropna(how="all")
```
